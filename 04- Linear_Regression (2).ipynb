## Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

# Load CSV File with Pandas

```python
# Load the dataset
file_path = '/gdrive/MyDrive/data/IndiaPopulation_2021.csv'
df = pd.read_csv(file_path)

# Display the first few rows
df.head()
```

## Visualize the Data

```python
# Plotting the data
plt.scatter(df['Year'], df['Population'])
plt.title('India Population Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

## Fitting the Linear Regression Model


```python
# Reshape the data
X = df['Year'].values.reshape(-1, 1)
y = df['Population'].values

# Initialize the model
model = LinearRegression()

# Fit the model
model.fit(X, y)

# Get the slope (m) and intercept (c)
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (m): {slope}")
print(f"Intercept (c): {intercept}")
```

## Plotting the Regression Line


```python
# Predict population
y_pred = model.predict(X)

# Plotting the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.title('Linear Regression: India Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

```python
# Calculating Mean Squared Error and R-squared
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

## Make Predictions

```python
# Predict population for the year 2025
year = np.array([[2025]])
predicted_population = model.predict(year)
print(f"Predicted Population for 2025: {predicted_population[0]}")
```
