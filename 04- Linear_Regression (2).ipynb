## Import Required Libraries

# Load CSV File with Pandas

```python
# Load the dataset
file_path = '/gdrive/MyDrive/data/IndiaPopulation_2021.csv'
df = pd.read_csv(file_path)

# Display the first few rows
df.head()
```

## Visualize the Data

```python
# Plotting the data
plt.scatter(df['Year'], df['Population'])
plt.title('India Population Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

## Fit the Linear Regression Model
.reshape(-1, 1)

Ensures the array is 2-dimensional, as expected by LinearRegression.

-1 means “infer the number of rows automatically”

1 means there is 1 feature column (here, the Year)

```python
# Reshape the data
X = df['Year'].values.reshape(-1, 1)
y = df['Population'].values

# Initialize the model
model = LinearRegression()

# Fit the model
model.fit(X, y)

# Get the slope (m) and intercept (c)
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (m): {slope}")
print(f"Intercept (c): {intercept}")
```

## Plot the Regression Line
Slope (m): Indicates the rate of population growth per year.

Intercept (c): Represents the estimated population at year 0 (which may not be realistic but serves as a mathematical reference).

R-squared: Measures the proportion of variance in the dependent variable predictable from the independent variable. A higher R² indicates a better fit.

```python
# Predict population
y_pred = model.predict(X)

# Plotting the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.title('Linear Regression: India Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

## Evaluate the Model
**What R² Means**

R² = 1: Perfect fit; the regression line explains 100% of the variation in the data.

R² = 0: The model explains none of the variation; the line is no better than just predicting the mean.

0 < R² < 1: The proportion of variance in
𝑌
Y explained by
𝑋
X.

R² < 0: The model is worse than just predicting the mean (rare but possible with poor fits).

image.png

```python
# Calculate Mean Squared Error and R-squared
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

## Make Predictions

```python
# Predict population for the year 2025
year = np.array([[2025]])
predicted_population = model.predict(year)
print(f"Predicted Population for 2025: {predicted_population[0]}")
```
